```python
# 导入json模块
import json
```

```python
# 定义训练集json文件名
train_file = "easy_train.json"
# 定义删除后的json文件名
new_train_file = "new_easy_train.json"
```

```python
# 打开并读取训练集json文件，获取images字段、categories字段和annotations字段
with open(train_file, 'r') as f:
    json_data = json.load(f)
images = json_data["images"]
categories = json_data["categories"]
annotations = json_data["annotations"]
```

```python
images
```

```python
# 定义想要删除的照片的名称或id，假设是"test.jpg"或者1
image_name = "200004148_00074_2.jpg"
image_id = 954
```

```python
# 使用列表推导式，从images字段和annotations字段中删除属于这张照片的部分，根据照片名称或id进行匹配
new_images = [image for image in images if image["file_name"] != image_name and image["id"] != image_id]
new_annotations = [anno for anno in annotations if anno["image_id"] != image_id]
# 打开并写入删除后的训练集文件
with open(new_train_file, 'w') as f:
    json.dump({"images": new_images, "categories": categories, "annotations": new_annotations}, f)
```
